## **Getting started**

## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

## Machine learning
</div>

With Colab you can import an image dataset, train an image classifier on it, and evaluate the model, all in just [a few lines of code](https://colab.research.google.com/github/tensorflow/docs/blob/master/site/en/tutorials/quickstart/beginner.ipynb). Colab notebooks execute code on Google's cloud servers, meaning you can leverage the power of Google hardware, including [GPUs and TPUs](#using-accelerated-hardware), regardless of the power of your machine. All you need is a browser.

```
import numpy as np 
import cv2
import matplotlib.pyplot as plt 

# c nh
img = cv2.imread('test.tif',0)
m,n = img.shape[:2]

# Thm nhiu Gaussian vo nh img
gia_tri_TB = 10
phuong_sai = 25
noise = np.random.normal(loc=gia_tri_TB,scale=phuong_sai,size=(m,n))
Gau_noisy_img = img + noise

# Thm nhiu Rayleigh vo nh img
phuong_sai = 40
noise = np.random.rayleigh(scale=phuong_sai,size=(m,n))
Ray_noisy_img = img + noise

# Thm nhiu Erlang (Gammar) vo nh img
K = 2.0
phuong_sai = 18
noise = np.random.gamma(shape=K,scale=phuong_sai,size=(m,n))
Gam_noisy_img = img + noise

# Thm nhiu hm m vo nh img
phuong_sai = 26
noise = np.random.exponential(scale=phuong_sai,size=(m,n))
Exp_noisy_img = img + noise

# Thm nhiu Uniform vo nh img
a,b = 10,100
noise = np.random.uniform(low=a,high=b,size=(m,n))
Uni_noisy_img = img + noise

# Thm nhiu mui tiu (add salt and pepper) vo nh img
number_black = int(m*n*0.05)  # nh ngha s im en
number_white = int(m*n*0.05)  # nh ngha s im trng
# Ly gi tr nguyn ngu nhin trong on 0..m
# Gi tr ny s biu din ta  im en theo hng
m_blacks = np.random.randint(0,m,number_black)
# Ly gi tr nguyn ngu nhin trong on 0..n
# Gi tr ny s biu din ta  im en theo ct
n_blacks = np.random.randint(0,n,number_black)
# Ly gi tr nguyn ngu nhin trong on 0..m
# Gi tr ny s biu din ta  im trng theo hng
m_whites = np.random.randint(0,m,number_white)
# Ly gi tr nguyn ngu nhin trong on 0..n
# Gi tr ny s biu din ta  im trng theo ct
n_whites = np.random.randint(0,n,number_white)

SP_noisy_img = np.copy(img) # Sao chp nh img  to nh SP_noisy_img
# Thit lp mc xm = 0 (im en) cho im nh c ta  (m_blacks,n_blacks)
SP_noisy_img[m_blacks,n_blacks] = 0
# Thit lp mc xm = 255 (im trng) cho im nh c ta  (m_whites,n_whites)
SP_noisy_img[m_whites,n_whites] = 255

# 1. Hin th nh gc, nh nhiu v histogram
# To ca s 1  hin th nh cho nhiu Gaussian
fig1 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
#To 9 vng v con
(ax1, ax2), (ax3, ax4) = fig1.subplots(2, 2)
# Hin th nh gc
ax1.imshow(img, cmap='gray')
ax1.set_title('nh gc')
ax1.axis('off')
# Hin th histogram nh gc
ax2.hist(img.flatten(),bins=256)
ax2.set_title('Histogram')
# Hin th nh nhiu Gaussian
ax3.imshow(Gau_noisy_img, cmap='gray')
ax3.set_title('nh nhiu Gaussian')
ax3.axis('off')
# Hin th histogram nh nhiu gaussian
ax4.hist(Gau_noisy_img.flatten(),bins=256)
ax4.set_title('Hitogram')
plt.show()

# 2. To ca s 2  hin th nh cho nhiu Rayleigh
fig2 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
#To 9 vng v con
(ax1, ax2), (ax3, ax4) = fig2.subplots(2, 2)
# Hin th nh gc
ax1.imshow(img, cmap='gray')
ax1.set_title('nh gc')
ax1.axis('off')
# Hin th histogram nh gc
ax2.hist(img.flatten(),bins=256)
ax2.set_title('Hitogram')
# Hin th nh nhiu Rayleigh
ax3.imshow(Ray_noisy_img, cmap='gray')
ax3.set_title('nh nhiu Rayleigh')
ax3.axis('off')
# Hin th histogram nh nhiu Rayleigh
ax4.hist(Ray_noisy_img.flatten(),bins=256)
ax4.set_title('Histogram')
plt.show()

# 3. To ca s 3  hin th nh cho nhiu Erlang (Gamma)
fig2 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
#To 9 vng v con
(ax1, ax2), (ax3, ax4) = fig2.subplots(2, 2)
# Hin th nh gc
ax1.imshow(img, cmap='gray')
ax1.set_title('nh gc')
ax1.axis('off')
# Hin th histogram nh gc
ax2.hist(img.flatten(),bins=256)
ax2.set_title('Hitogram')
# Hin th nh nhiu Erlang (Gammar)
ax3.imshow(Gam_noisy_img, cmap='gray')
ax3.set_title('nh nhiu Erlang (Gammar)')
ax3.axis('off')
# Hin th histogram nh nhiu Erlang (Gammar)
ax4.hist(Gam_noisy_img.flatten(),bins=256)
ax4.set_title('Histogram')
plt.show()

# 4. To ca s 4  hin th nh cho nhiu nhiu hm m
fig2 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
#To 4 vng v con
(ax1, ax2), (ax3, ax4) = fig2.subplots(2, 2)
# Hin th nh gc
ax1.imshow(img, cmap='gray')
ax1.set_title('nh gc')
ax1.axis('off')
# Hin th histogram nh gc
ax2.hist(img.flatten(),bins=256)
ax2.set_title('Hitogram')
# Hin th nh nhiu nhiu hm m
ax3.imshow(Exp_noisy_img, cmap='gray')
ax3.set_title('nh nhiu hm m')
ax3.axis('off')
# Hin th histogram nh nhiu nhiu hm m
ax4.hist(Exp_noisy_img.flatten(),bins=256)
ax4.set_title('Histogram')
plt.show()

# 5. To ca s 5  hin th nh cho nhiu nhiu Uniform
fig2 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
#To 4 vng v con
(ax1, ax2), (ax3, ax4) = fig2.subplots(2, 2)
# Hin th nh gc
ax1.imshow(img, cmap='gray')
ax1.set_title('nh gc')
ax1.axis('off')
# Hin th histogram nh gc
ax2.hist(img.flatten(),bins=256)
ax2.set_title('Hitogram')
# Hin th nh nhiu nhiu Uniform
ax3.imshow(Uni_noisy_img, cmap='gray')
ax3.set_title('nh nhiu Uniform')
ax3.axis('off')
# Hin th histogram nh nhiu Uniform
ax4.hist(Uni_noisy_img.flatten(),bins=256)
ax4.set_title('Histogram')
plt.show()

# 6. To ca s 6  hin th nh cho nhiu nhiu mui tiu (nhiu xung)
fig2 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
#To 4 vng v con
(ax1, ax2), (ax3, ax4) = fig2.subplots(2, 2)
# Hin th nh gc
ax1.imshow(img, cmap='gray')
ax1.set_title('nh gc')
ax1.axis('off')
# Hin th histogram nh gc
ax2.hist(img.flatten(),bins=256)
ax2.set_title('Hitogram')
# Hin th nh nhiu nhiu mui tiu (nhiu xung)
ax3.imshow(SP_noisy_img, cmap='gray')
ax3.set_title('nh nhiu mui tiu')
ax3.axis('off')
# Hin th histogram nh nhiu mui tiu (nhiu xung)
ax4.hist(SP_noisy_img.flatten(),bins=256)
ax4.set_title('Histogram')
plt.show()
```
